```python
import numpy as np
```

```python
N = int(input())
Y = np.array([float(i) for i in input().split()])
```

```python
fucking_func = lambda x: 8*np.sin(x) + 7*x + 5
```

```python
iterator = 0.0
while np.all(round(fucking_func(iterator),3) != round(Y[0],3)):
    iterator+=0.1
    print(fucking_func(iterator), '    ', Y[0], '\n')
print(iterator)
```

```python
# step = 2*10**8/N
# X = np.arange(-10**(8), 10**(8), step)
X = np.arange(N)
X
```

```python
A = np.array([[i**2 for i in X[-3:]], X[-3:], [1, 1, 1]]).transpose()
det = np.linalg.det(A)

coef_a = np.copy(A)
coef_b = np.copy(A)
coef_c = np.copy(A)

coef_a[:,0] = Y[-3:]
coef_b[:,1] = Y[-3:]
coef_c[:,2] = Y[-3:]

det_a = np.linalg.det(coef_a)
det_b = np.linalg.det(coef_b)
det_c = np.linalg.det(coef_c)

a, b, c = (det_a/det, det_b/det, det_c/det)

y = lambda x: x**2*a + x*b + c

y(N)
```
