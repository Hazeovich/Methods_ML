```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Lasso
```

```python
def split_params(line: str):
    line_split = line.strip().split()
    return int(line_split[0]), int(line_split[1]), int(line_split[2]), float(line_split[3])

with open('data.txt') as file:
    data_row = file.readlines()
file.close()

row_counter = 373
TEST_COUNT = int(data_row[0].strip())
n, m, k, s = split_params(data_row[row_counter])
current_example = []
for i in range(n):
    current_example.append(data_row[i+row_counter+1].strip().split())
```

```python
cols = [f'x{i+1}' for i in range(m)]
cols.append('y')
df = pd.DataFrame(current_example, columns=cols)
df
```

```python
k_search = k + 1
alpha = 0.00001

while k_search != k:
    clf = Lasso(alpha=alpha)
    clf.fit(df.iloc[:, :-1], df.iloc[:,-1])
    k_search = m - len(np.nonzero(clf.coef_)[0])
    if k_search < k:
        alpha *= 5
    elif k_search > k:
        alpha *= 0.5
print(f"need:{k}, search:{k_search}\n\n\n\n")
print(*clf.coef_, clf.intercept_)
```
