```python
import pandas as pd
df = pd.read_csv('fillna_data.csv', index_col='date')
df_out = pd.read_csv('fillna_data_out.csv', index_col='date')
df
```

```python
import pandas as pd

def fillna_date(data: pd.DataFrame, function: str = 'mean') -> pd.DataFrame:
    
    month_per = pd.PeriodIndex(data.index, freq="M")                            #Получаем список месяцов длиной = длине data
    year_per = pd.PeriodIndex(data.index, freq="Y")                             #Получает список годов длиной = длине data
    
    _data = data.groupby(month_per).transform(function)                         #Заполняем пропуски по месяцам
    _data.fillna(data.groupby(year_per).transform(function), inplace=True)      #Заполняем пропуски по годам
    _data.fillna(data.agg(function), inplace=True)                              #Заполняем пропуски по всей колонке
       
    return data.fillna(_data)
```

```python
b = fillna_date(df)
b
```

```python
(df_out - b).apply(lambda x: x != 0.0).value_counts()
```
